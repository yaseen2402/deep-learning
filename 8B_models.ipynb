```python
from unsloth import FastLanguageModel
from transformers import AutoTokenizer, pipeline
```

```python
# See more at https://huggingface.co/unsloth
model_name = "unsloth/gemma-3-12b-it"
# model_name = "unsloth/codellama-13b-bnb-4bit"


# ✅ Step 4: Load tokenizer
tokenizer = AutoTokenizer.from_pretrained(model_name, use_fast=True)

# ✅ Step 5: Load Unsloth model with 4-bit support
import torch
model, tokenizer = FastLanguageModel.from_pretrained(
    model_name = model_name,
    max_seq_length = 16384,
    dtype = torch.float16,             # Auto-detects bf16/fp16/cuda
    load_in_4bit = True,      # You can change to False for full-precision
)

# ✅ Step 6: Compile model (optional, speeds up inference)
FastLanguageModel.for_inference(model)
```

```python
# ✅ Step 7: Define your test prompt
prompt = """### Instruction:
can you write a program to fine the factorial of a number in java
### Response:
"""


# ✅ Step 8: Run inference
pipe = pipeline("text-generation", model=model, tokenizer=tokenizer)

outputs = pipe(prompt, max_new_tokens=1000, temperature=0.7, do_sample=True)
print(outputs[0]["generated_text"])
```
